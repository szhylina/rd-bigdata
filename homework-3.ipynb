```pyspark
# шлях до файлів з даними
from pyspark.sql.types import *
s3_bucket_name = 's3a://robot-dreams-source-data/home-work-1/nyc_taxi/'
```

```pyspark
# вичитування файлів з даними по зелених таксі із заданою схемою
from functools import reduce
from pyspark.sql import DataFrame
from pyspark.sql.functions import col

schema_green_taxi = StructType([
    StructField("VendorID", LongType(), True),
    StructField("lpep_pickup_datetime", TimestampType(), True),
    StructField("lpep_dropoff_datetime", TimestampType(), True),
    StructField("store_and_fwd_flag", StringType(), True),
    StructField("RatecodeID", LongType(), True),
    StructField("PULocationID", LongType(), True),
    StructField("DOLocationID", LongType(), True),
    StructField("passenger_count", LongType(), True),
    StructField("trip_distance", DoubleType(), True),
    StructField("fare_amount", DoubleType(), True),
    StructField("extra", DoubleType(), True),
    StructField("mta_tax", DoubleType(), True),
    StructField("tip_amount", DoubleType(), True),
    StructField("tolls_amount", DoubleType(), True),
    StructField("ehail_fee", IntegerType(), True),
    StructField("improvement_surcharge", DoubleType(), True),
    StructField("total_amount", DoubleType(), True),
    StructField("payment_type", LongType(), True),
    StructField("trip_type", DoubleType(), True),
    StructField("congestion_surcharge", DoubleType(), True)
])

green_taxi_data_url = f"{s3_bucket_name}green/*"
print(f"Using Green Taxi data from: {green_taxi_data_url}")
green_taxi_dfs = []

for file in spark.read.parquet(green_taxi_data_url).inputFiles():
    denormalized_df = spark.read.parquet(file)
    green_taxi_dfs.append(
            denormalized_df.select(
            [col(field.name).cast(field.dataType).alias(field.name)
            for field in schema_green_taxi.fields]
        )
    )

green_taxi_df = reduce(DataFrame.union, green_taxi_dfs)
print(f"Green Taxi data loaded. Records: {green_taxi_df.count()}")
green_taxi_df.printSchema()
```

```pyspark
# вичитування файлів з даними по жовтих таксі із заданою схемою

schema_yellow_taxi = StructType([
    StructField("VendorID", LongType(), True),
    StructField("tpep_pickup_datetime", TimestampType(), True),
    StructField("tpep_dropoff_datetime", TimestampType(), True),
    StructField("store_and_fwd_flag", StringType(), True),
    StructField("RatecodeID", LongType(), True),
    StructField("PULocationID", LongType(), True),
    StructField("DOLocationID", LongType(), True),
    StructField("passenger_count", LongType(), True),
    StructField("trip_distance", DoubleType(), True),
    StructField("fare_amount", DoubleType(), True),
    StructField("extra", DoubleType(), True),
    StructField("mta_tax", DoubleType(), True),
    StructField("tip_amount", DoubleType(), True),
    StructField("tolls_amount", DoubleType(), True),
    StructField("improvement_surcharge", DoubleType(), True),
    StructField("total_amount", DoubleType(), True),
    StructField("payment_type", LongType(), True),
    StructField("congestion_surcharge", DoubleType(), True),
    StructField("airport_fee", IntegerType(), True)
])

yellow_taxi_data_url = f"{s3_bucket_name}yellow/*"
print(f"Using Yellow Taxi data from: {yellow_taxi_data_url}")
yellow_taxi_dfs = []

for file in spark.read.parquet(yellow_taxi_data_url).inputFiles():
    denormalized_df = spark.read.parquet(file)
    yellow_taxi_dfs.append(
            denormalized_df.select(
            [col(field.name).cast(field.dataType).alias(field.name)
            for field in schema_yellow_taxi.fields]
        )
    )
    
yellow_taxi_df = reduce(DataFrame.union, yellow_taxi_dfs)
print(f"Yellow Taxi data loaded. Records: {yellow_taxi_df.count()}")
yellow_taxi_df.printSchema()
```

```pyspark
# додаємо колонку з тривалістю поїздок у хвилинах та колонку з кольором таксі
from pyspark.sql.functions import lit, unix_timestamp

green_taxi_new_columns_df = green_taxi_df.withColumn('duration_min', (unix_timestamp('lpep_dropoff_datetime') - unix_timestamp('lpep_pickup_datetime')) / 60).withColumn('taxi_type', lit('green'))

yellow_taxi_new_columns_df = yellow_taxi_df.withColumn('duration_min', (unix_timestamp('tpep_dropoff_datetime') - unix_timestamp('tpep_pickup_datetime')) / 60).withColumn('taxi_type', lit('yellow'))
```

```pyspark
# вилучаємо поїздки з відстанню < 0.1 км (дані в милях, 0.1 км ≈ 0.062 милі), тарифом < $2, тривалістю < 1 хв.
green_taxi_filtered_df = green_taxi_new_columns_df.filter(col('trip_distance') > 0.062).filter(col('fare_amount') >= 2).filter(col('duration_min') >= 1)
yellow_taxi_filtered_df = yellow_taxi_new_columns_df.filter(col('trip_distance') > 0.062).filter(col('fare_amount') >= 2).filter(col('duration_min') >= 1)
```

```pyspark
# перейменовуємо стовпчики в меншому датасеті та об'єднуємо їх

green_taxi_reshaped_df = green_taxi_filtered_df.withColumnsRenamed({'lpep_pickup_datetime': 'tpep_pickup_datetime', 'lpep_dropoff_datetime': 'tpep_dropoff_datetime'})

raw_trips_df = yellow_taxi_filtered_df.unionByName(green_taxi_reshaped_df, allowMissingColumns=True)
```

```pyspark
# додаємо нові стопвчики з даними  pickup_hour — годину початку поїздки, pickup_day_of_week — день тижня
from pyspark.sql.functions import hour, dayofweek

raw_trips_df_new_columns = raw_trips_df.withColumn('pickup_hour', hour('tpep_pickup_datetime')).withColumn('pickup_day_of_week', dayofweek('tpep_pickup_datetime'))
```

```pyspark
# читаємо дані про локації з csv файла
schema_csv = StructType([
    StructField("LocationID", LongType(), True),
    StructField("Borough", StringType(), True),
    StructField("Zone", StringType(), True),
    StructField("service_zone", StringType(), True)
])
df_csv = spark.read.option("header", True).schema(schema_csv).csv(f"{s3_bucket_name}taxi_zone_lookup.csv")
```

```pyspark
from pyspark.sql.functions import broadcast

raw_trips_final_df = raw_trips_df_new_columns.join(
    broadcast(df_csv), raw_trips_df_new_columns.PULocationID == df_csv.LocationID, how="left_outer"
).drop('LocationID', 'Borough', 'service_zone').withColumnRenamed('Zone', 'pickup_zone')

raw_trips_final_df.show(5)

raw_trips_final_df = raw_trips_final_df.join(
    broadcast(df_csv), raw_trips_df_new_columns.DOLocationID == df_csv.LocationID, how="left_outer"
).drop('LocationID', 'Borough', 'service_zone').withColumnRenamed('Zone', 'dropoff_zone')

raw_trips_final_df.show(5)
```

```pyspark
# формуємо датафрейм zone_summary з агрегованими даними по pickup_zone
from pyspark.sql.functions import count, avg, max, min, sum, when

zone_summary = raw_trips_final_df.groupBy("pickup_zone").agg(
    count("*").alias("total_trips"),
    avg("trip_distance").alias("avg_trip_distance"),
    avg("total_amount").alias("avg_total_amount"),
    avg("tip_amount").alias("avg_tip_amount"),
    max("trip_distance").alias("max_trip_distance"),
    min("tip_amount").alias("min_tip_amount"),
    sum(when(col("taxi_type") == "raw_trips_final_df", 1).otherwise(0)).alias("yellow_count"),
    sum(when(col("taxi_type") == "green", 1).otherwise(0)).alias("green_count")
)

zone_summary = zone_summary.withColumn("yellow_share", col("yellow_count") / col("total_trips"))
zone_summary = zone_summary.withColumn("green_share", col("green_count") / col("total_trips"))

zone_summary = zone_summary.drop("yellow_count", "green_count")

zone_summary.show(5)
```

```pyspark
# збегігаємо агреговані дані в особистий s3
from datetime import datetime

date_str = datetime.today().strftime('%Y-%m-%d')
s3_personal_bucket_name_zs = f"s3a://szhylina-hw/results/zone_statistic/{date_str}/zone_statistic.parquet"

zone_summary.write.mode("overwrite").parquet(s3_personal_bucket_name_zs)
```

```pyspark
# формуємо датафрейм zone_days_summary з агрегованими даними по днях тижня та зонах

zone_days_summary = raw_trips_final_df.groupBy("pickup_zone", "pickup_day_of_week").agg(
    count("*").alias("total_trips"),
    (sum(when(col("fare_amount") > 30, 1).otherwise(0)) / count("*")).alias("high_fare_share")
)

zone_days_summary.show(5)
```

```pyspark
# зберігаємо агреговані дані в особистий s3

s3_personal_bucket_name_zds = f"s3a://szhylina-hw/results/zone_days_statstic/{date_str}/zone_days_statstic.parquet"
zone_days_summary.write.mode("overwrite").parquet(s3_personal_bucket_name_zds)
```
